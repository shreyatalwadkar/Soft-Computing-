```python
import numpy as np

# Training data (bipolar inputs)
inputs = np.array([[1, 1],
                   [1, -1],
                   [-1, 1],
                   [-1, -1]])

# Target outputs (for AND gate in bipolar form)
targets = np.array([1, -1, -1, -1])

# Initialize weights
weights = np.zeros(inputs.shape[1])
bias = 0

# Hebbian Learning
for i in range(len(inputs)):
    weights = weights + inputs[i] * targets[i]
    bias = bias + targets[i]

print("Final Weights:", weights)
print("Final Bias:", bias)
```
