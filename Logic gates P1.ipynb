AND gate usage 

```python
def mp_and(x1, x2):
    w1, w2 = 1, 1
    theta = 2
    
    summation = x1*w1 + x2*w2
    
    if summation >= theta:
        return 1
    else:
        return 0

# Testing
print("AND Gate using McCulloch-Pitts Model")
for a in [0,1]:
    for b in [0,1]:
        print(a, b, "->", mp_and(a,b))
```

OR gate usage 

```python
def mp_or(x1, x2):
    w1, w2 = 1, 1
    theta = 1
    
    summation = x1*w1 + x2*w2
    
    if summation >= theta:
        return 1
    else:
        return 0
```

NOT gate usage 

```python
def mp_not(x):
    w = -1
    theta = 0
    
    summation = x*w
    
    if summation >= theta:
        return 1
    else:
        return 0

# Testing NOT
print("\nNOT Gate using McCulloch-Pitts Model")
for a in [0,1]:
    print(a, "->", mp_not(a))
```
